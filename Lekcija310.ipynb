## Nikola Tesla

**Nikola Tesla** je najpoznatiji po svom doprinosu u projektovanju modernog sistema napajanja *naizmeničnom strujom*. 

Njegovi najznačajniji pronalasci su:
- polifazni sistem
- obrtno magnetsko polje
- ansinhroni motor
- sinhroni motor
- Teslin transformator
